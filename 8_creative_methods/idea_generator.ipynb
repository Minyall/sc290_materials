```python

from random import choice



def idea_me(my_topic:str, 
            my_method:str = None,
            segmented_by:str=None,
            level_of_analysis:str=None):
    
    method_choices = [
    "Topic Modelling",
    "TFIDF Words",
    "Sentiment Analysis",
    "Entity Recognition",
    "Network Communities", # Which nodes are closer together
    "Network Significance" # Which nodes are 'important' by metric (degree, pagerank, eigenvector etc)
]
    segmented_choices = [
    "Time period",
    "Section",
    "Tag",
    "Entity presence",
    "Topic",
    "Entity Community"
]
    level_choices = ['Whole Corpus',
                      'Subset',
                        'Document',
                        'Paragraph']
    
    if my_method is None:
        my_method = choice(method_choices)

    if segmented_by is None:
        segmented_by = choice(segmented_choices)
    if level_of_analysis is None:
        level_of_analysis = choice(level_choices)

    print(f'Topic: {my_topic}')
    print(f'Main Method: {my_method}')
    print(f'Segmented by: {segmented_by}')
    print(f'Exploring the data at the {level_of_analysis} level')


    


```

```python
idea_me('The far right')
```
